```python
from connection import *
import pandas as pd
import sqlite3

URL_TOP_RATED_MOVIE = "https://api.themoviedb.org/3/movie/top_rated?language=en-US&page=1"
URL_TOP_RATED_SERIES = "https://api.themoviedb.org/3/tv/top_rated?language=en-US&page=1"
URL_GENRE_MOVIES_LIST = "https://api.themoviedb.org/3/genre/movie/list"

SERIE_ID = 1668
url_serie_details = f"https://api.themoviedb.org/3/tv/{SERIE_ID}?language=en-US"

key = Tmdb_api_key()

url1 = URL_TOP_RATED_MOVIE
url2 = URL_TOP_RATED_SERIES
url3 = url_serie_details
url4 = URL_GENRE_MOVIES_LIST

response1 = get_request(url1,key.get_headers())
response2 = get_request(url2,key.get_headers())
response3 = get_request(url3,key.get_headers())
response4 = get_request(url4,key.get_headers())

print(response1.text)
print(response2.text)
print(response3.text)
print(response4.text)
```

```python
def getDataFrameFromList(json_list):
    adult = [item['adult'] for item in json_list]
    original_title = [item['original_title'] for item in json_list]
    original_language = [item['original_language'] for item in json_list]
    overview = [item['overview'] for item in json_list]
    popularity = [item['popularity'] for item in json_list]
    release_date = [item['release_date'] for item in json_list]

    return pd.DataFrame({
        "adult": adult,
        'orignlan_title': original_title,
        'original_language':original_language,
        'overview':overview,
        'popularity':popularity,
        'release_date':release_date
    })
```

```python
topratedmoviesjson = response1.json()['results']
topratedseriresjson = response1.json()['results']
moviegenresjson = response4.json()['genres']

dftopratedmovies = getDataFrameFromList(topratedmoviesjson)
dftopratedseries = getDataFrameFromList(topratedseriresjson)
dfmoviegenres = pd.DataFrame(moviegenresjson)

conn = sqlite3.connect('themoviedb.db')
dftopratedmovies.to_sql('topratedmovies', conn, if_exists='replace', index=False)
dftopratedseries.to_sql('topratedseries', conn, if_exists='replace', index=False)
dfmoviegenres.to_sql('genres', conn, if_exists='replace', index=False)

conn = sqlite3.connect('themoviedb.db')
query  = "SELECT name FROM sqlite_master WHERE type='table'"
df = pd.read_sql_query(query, conn)

print(df)

conn.close()

```
