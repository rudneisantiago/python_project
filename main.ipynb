```python
from api_connection import *
from db_connection import *
from utils import *
import pandas as pd
import sqlite3

URL_TOP_RATED_MOVIE = "https://api.themoviedb.org/3/movie/top_rated?language=en-US&page=1"
URL_TOP_RATED_SERIES = "https://api.themoviedb.org/3/tv/top_rated?language=en-US&page=1"
URL_GENRE_MOVIES_LIST = "https://api.themoviedb.org/3/genre/movie/list"

SERIE_ID = 1668
url_serie_details = f"https://api.themoviedb.org/3/tv/{SERIE_ID}?language=en-US"

key = Tmdb_api_key()

url1 = URL_TOP_RATED_MOVIE
url2 = URL_TOP_RATED_SERIES
url3 = url_serie_details
url4 = URL_GENRE_MOVIES_LIST

response1 = get_request(url1,key.get_headers())
response2 = get_request(url2,key.get_headers())
response3 = get_request(url3,key.get_headers())
response4 = get_request(url4,key.get_headers())
```

```python
topratedmoviesjson = response1.json()['results']
topratedseriresjson = response1.json()['results']
moviegenresjson = response4.json()['genres']

dftopratedmovies = get_dataframe_from_list(topratedmoviesjson)
dftopratedseries = get_dataframe_from_list(topratedseriresjson)
dfmoviegenres = pd.DataFrame(moviegenresjson)

db_name = 'themoviedb.db'
table1_name = "topratedmovies"
table2_name = "topratedseries"
table3_name = "genres" 

conn = open_connection(db_name)
df_to_sql(dftopratedmovies,table1_name,conn)
df_to_sql(dftopratedseries,table2_name,conn)
df_to_sql(dfmoviegenres,table3_name,conn)

df_tables = get_all_tables(conn)

print(df_tables)
```

```python
df_movies = get_all_from_table(table1_name,conn)
df_series = get_all_from_table(table2_name,conn)
df_genres = get_all_from_table(table3_name,conn)

display(df_movies.info())
display(df_series.info())
display(df_genres.info())

display(df_movies)
display(df_series)
display(df_genres)

close_connection(conn)
```
